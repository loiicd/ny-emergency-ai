```python
INICENTS_QUANTITY = 3
```

```python
import joblib
from pathlib import Path

model = joblib.load(f'{Path.cwd()}/../models/model.pkl')
features = joblib.load(f'{Path.cwd()}/../models/features.pkl')
target_attribute = joblib.load(f'{Path.cwd()}/../models/target.pkl')

dummy_features = model.feature_names_in_
```

```python
import pandas as pd
from services.createFictionalData import createFictionalData

data = createFictionalData(INICENTS_QUANTITY, features)

df = pd.DataFrame(data)

df.head(20)
```

```python
df = pd.get_dummies(df, columns=features, dtype=int)

df.head()
```

```python
for dummy_feature in dummy_features:
  if dummy_feature not in df.columns:
    df[dummy_feature] = 0

df.head()
```

```python
df = df[dummy_features]

df.head()
```

```python
predictions = model.predict(df)

print('Predictions:', predictions)
```

```python
predictions = predictions.round()

print('Rounded Predictions:', predictions)
```
