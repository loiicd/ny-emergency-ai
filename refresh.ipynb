```python
ITERATIONS = 20
TEST_SIZE = 0.3
FEATURES = ['alarm_box_borough', 'incident_classification']
TARGET_ATTRIBUTE = 'incident_response_seconds_qy'
SAVE_AS = 'version-01'
```

```python
import pandas as pd

df = pd.read_json('/Users/loic.doerr/dev/ny-emergency-ai/data/fire-incidents/interims.json')
df.dropna(inplace=True)

df.head()
```

```python
features = df[FEATURES]
features = pd.get_dummies(features, columns=FEATURES, dtype=int)

features.head()
```

```python
target = df[TARGET_ATTRIBUTE]

target.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics

models = []
errors = []

for iteration in range(ITERATIONS):
  features_train, features_test, target_train, target_test = train_test_split(features, target, test_size=TEST_SIZE, random_state=iteration) 
  model = LinearRegression()
  model.fit(features_train, target_train)
  target_pred = model.predict(features_test)
  mse = metrics.mean_squared_error(target_test, target_pred)
  models.append(model)
  errors.append(mse)

print('Errors', errors)
```

```python
import matplotlib.pyplot as plt
import numpy as np

average_error = np.mean(errors)

plt.plot(errors)
plt.axhline(average_error, color='r', linestyle='--', label='Average MSE')
plt.xlabel('Model Index')
plt.ylabel('MSE')
plt.title('MSE of Linear Regression Models')
plt.legend()
plt.show()
```

```python
best_model_index = errors.index(min(errors))
best_model = models[best_model_index]

print('Best Model MSE', min(errors))
```

```python
import joblib

joblib.dump(best_model, f'./models/model.pkl')

joblib.dump(FEATURES, './models/features.pkl')
joblib.dump(TARGET_ATTRIBUTE, './models/target.pkl')
```
