```python
import joblib

model = joblib.load('./models/model.pkl')
features = joblib.load('./models/features.pkl')
target_attribute = joblib.load('./models/target.pkl')

dummy_features = model.feature_names_in_

print('Features:', features)
print('Dummy Features:', dummy_features)
print('Target Attribute:', target_attribute)
```

```python
import pandas as pd

df = pd.DataFrame(data)

df.head()
```

```python
df = pd.get_dummies(df, columns=features, dtype=int)

df.head()
```

```python
for dummy_feature in dummy_features:
  if dummy_feature not in df.columns:
    df[dummy_feature] = 0

df.head()
```

```python
df = df[dummy_features]

df.head()
```

```python
predictions = model.predict(df)

print('Predictions:', predictions)
```

```python
predictions = predictions.round()

print('Rounded Predictions:', predictions)
```
